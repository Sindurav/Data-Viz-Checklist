```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df_g=df.groupby("species").sum().reset_index()
df_g
```

## Bar Plot
Used for comparision between the categories of data.

```python
df_g.plot(kind='bar',x='species',y='petal_length')
```

## Histogram
Used when we need to plot the frequency of score occurences in the continous data.

```python
df.hist(grid=False,bins=5,column='petal_length',color="#41817F")
```

## Distplot
Used when we need to examine the distibution of the variable along with its frequency score. This is used as univariate distribution analysis.

```python
sns.distplot(df.sepal_length)
```

## Box Plot
Box plot is used when we need to examine the distribution of data using descriptive statistics like min, quartiles, max values

```python
sns.boxplot(data=df)
```

## Violin Plot
Violin plot is similar to Box plot. In addition they display probability density of the data at different values, usually smoothed by a kernel density estimator.

```python
sns.violinplot(x='species', y='petal_length', data=df)
```

## Scatter Plot
Scatter clearly depicts the relationship between two variables. It shows how one variable is affected by the other by plotting the correlation between them.

```python
df.plot(kind="scatter",x='sepal_length',y='petal_length')
```

## Bubble Chart
It is similar to Scatterplot, bubble chart is preferred when 3rd dimension is introduced in scatter plot. Here, size of the bubble describes the quantitative value of 3rd dimension data.

```python
plt.scatter(x='species', y='sepal_width', s='petal_length',marker="D", data=df_g,c="green", alpha=0.4, linewidth=6)
```

## Heatmap
Heatap is used to analyse the correlation between two varriable. We can also annot the values as shown below.

```python
# Calculate correlations
corr = df.corr()
 
# Heatmap
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

## Countplot
used when we need to analyse the count of variable in the data.

```python
sns.countplot(x='species', data=df)
```

## Pairplot
It is used to analyse both the distribution of single variable and relationship between two variables.

```python
sns.pairplot(df, hue="species", size=3)
```

## Parallel Co-Ordinates
It is most common approach to examine multivariate data. We can analyse the influence of multiple variables on a single target variable.

```python
import matplotlib.pyplot as plt
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(df, 'species', colormap=plt.get_cmap("Set2"))
plt.show()
```

## Squarify Plot
Squarify plot or treemap is used when we need to represent hirerchial data using nested rectangles.

```python
import squarify
squarify.plot(sizes=df_g['sepal_length'], label=df_g['species'], alpha=.8 )
plt.axis('off')
plt.show()
```

## Regplot
Regplot performs regression analysis and plots resulting regressiong line. Regplot has input parameter flexibility over lmplot (which is shown below) as it accepts numpy arrays and pandas series obeject.

```python
sns.regplot(x=df["sepal_length"], y=df["sepal_width"])
```

## Lmplot
Similar to regplot, performs regression analysis with regression line. In additional 'hue' parameter can be used here to differentiate the categories.

```python
sns.lmplot(x='sepal_length', y='petal_length', data=df,hue='species')
```

## Jointplot
Jointplot gives the scattered distribution along with the density curves in the spines.

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=df)
```
